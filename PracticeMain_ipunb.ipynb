по диагонали заполнены единицей. N – считывается с клавиатуры.

**Решение:**

```python
def numpy_2():


  def beautiful_matrix(N):
    arr = np.eye(N, N, 0, int)
    j = 0
    for i in np.flip(np.where(arr==1)[0]):
      arr[i][j] = 1
      j += 1
    print("\nРезультат:")
    for i in range(N):
      print(*arr[i])


  print("Введите размерность квадратной матрицы:")
  beautiful_matrix(int(input()))


numpy_2()
```

##**Задание №3**
**Задача:**
* Напишите функцию, которая создает прямоугольную матрицу A,
имеющую N строк и M столбцов (N,M считываются из файла "input.txt") со случайными числами. Вычесть из элементов каждой строки элемент этой строки с наименьшим значением. Создать прямоугольную матрицу B размерности NxM. Посчитать разность матриц C = B + А и вывести в файл "result.txt".

**Решение:**

```python
def numpy_3():


  def sum_matrix(A,B):
    matrix =  B + A
    return matrix


  def action(matrix):
    for i in range(matrix.shape[0]):
      minimum = np.min(matrix[i])
      for j in range(len(matrix[i])):
        matrix[i][j] -= minimum
    return matrix


  def create_matrix():
    with open('/content/drive/MyDrive/input.txt') as f:
      data = list(f.readline().split())
    N, M = int(data[0]), int(data[1])
    matrix = np.random.randint(1,10,(N,M))
    f.close()
    return matrix


  def output_matrix(message, matrix):
    with open('/content/drive/MyDrive/result.txt', 'a') as f:
      f.write(f'{message}\n')
      for i in range(matrix.shape[0]):
        f.write(f'{" ".join(map(str, matrix[i]))}\n')
      f.write('\n')
    f.close()


  with open('/content/drive/MyDrive/result.txt', 'w') as f:
    f.close()
  A = create_matrix()
  output_matrix("Матрица A:", A)
  A = action(A)
  output_matrix("Новая матрица A:", A)
  B = create_matrix()
  output_matrix("Матрица B:", B)
  C = sum_matrix(A,B)
  output_matrix("Матрица С:", C)
  print('Результат ждёт вас в файле "result.txt"')


numpy_3()
```

образом из 15 значений. При этом все элементы Series меньше среднего
арифметического заменить на заданное константное значение.

**Решение:**

```python
def create_series():
  constant = int(input("Введите константное значение: "))
  data = pd.Series(randint(1,100) for i in range(15))
  print(f'До замен:\n{data}\n\nСреднее арифметическое: {sum(data)//15}\n')
  data = pd.Series(i if i > sum(data)//15 else constant for i in data)
  print(f'После замен:\n{data}')


create_series()
```

* Функция отображения на одном рисунке диаграммы количества
спальных комнат в зависимости от района и диаграммы количества
ванных комнат в зависимости от района.

```python
def R_N():


  def Rooms_in_Neighborhood(message, j):
    Statistics = []
    Neighborhood_data = list(set(data['Neighborhood']))
    for i in Neighborhood_data:
      Statistics.append(len(data[(data['Neighborhood'] == i) & (data[message] == j) ]))
    return Statistics


  fig, (ax1, ax2) = plt.subplots(2,1,figsize=(10,10))
  area = set(data['Neighborhood'])
  x_axis = np.arange(len(area))
  width = 0.2
  for i in range(2):
    rooms_data = {}
    multiplier = 0
    if i == 0:
      message = 'Bedrooms'
      ax = ax1
    else:
      message = 'Bathrooms'
      ax = ax2
    quantity_rooms = set(data[message])
    for j in quantity_rooms:
      rooms_data[f'{j} rooms'] = Rooms_in_Neighborhood(message, j)
    for keys, values in rooms_data.items():
        offset = width * multiplier
        posts = ax.bar(x_axis + offset, values, width, label=keys)
        ax.bar_label(posts, padding=2)
        multiplier += 1
    ax.set_ylabel('Quantity')
    ax.set_xlabel('Neighborhoods')
    ax.set_title(f'Quantity {message} relatively Neighborhood')
    ax.set_xticks(x_axis + width, area)
    ax.legend(loc='upper left', ncols=len(rooms_data.keys()))
    ax.set_ylim(0, 5500 if i == 0 else 7000)
    plt.subplots_adjust(hspace=0.35)


  plt.show()


R_N()
```

**Анализ:**
*На диаграмме представлены два столбчатых графика, отображающих количество спален и ванных комнат в зависимости от типа района: сельский (Rural), пригород (Suburb) и городской (Urban).*
